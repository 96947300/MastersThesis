### Parkinson (1980) Daily variance

```r
# Daily Variance (Parkinson, 1980)

# united States

par(mfrow = c(3,1))

high <- Hi(GSPC)
low  <- Lo(GSPC)
nrow(high)
# Parkinson (1980) estimator for each day
sigma2_SP500 <- 0.361 * (log(high / low))^2

plot.xts(sigma2_SP500, col="black", grid.ticks.on = F)


# China
high <- Hi(`000001.SS`)
low  <- Lo(`000001.SS`)

# Parkinson (1980) estimator for each day
sigma2_SSEC <- 0.361 * (log(high / low))^2

plot.xts(sigma2_SSEC, col="black", grid.ticks.on = F)

# Indonesia
high <- Hi(JKSE)
low  <- Lo(JKSE)

# Parkinson (1980) estimator for each day
sigma2_JKSE <- 0.361 * (log(high / low))^2
sigma2_JKSE <- na.locf(sigma2_JKSE)

plot.xts(sigma2_JKSE, col="black", grid.ticks.on = F)

high <- Hi(NSEI)
low  <- Lo(NSEI)

# Parkinson (1980) estimator for each day
sigma2_NSEI <- 0.361 * (log(high / low))^2
sigma2_NSEI <- na.locf(sigma2_NSEI)

plot.xts(sigma2_NSEI, col="black", grid.ticks.on = F)

# Thailand
high <- Hi(SET.BK)
low  <- Lo(SET.BK)

# Parkinson (1980) estimator for each day
sigma2_SET.BK <- 0.361 * (log(high / low))^2
sigma2_SET.BK <- na.locf(sigma2_SET.BK)

plot.xts(sigma2_SET.BK, col="black", grid.ticks.on = F)

# Malaysia
high <- Hi(KLSE)
low  <- Lo(KLSE)

# Parkinson (1980) estimator for each day
sigma2_KLSE <- 0.361 * (log(high / low))^2
sigma2_KLSE <- na.locf(sigma2_KLSE)

plot.xts(sigma2_KLSE, col="black", grid.ticks.on = F)

# Philippines
high <- Hi(PSEI.PS)
low  <- Lo(PSEI.PS)

# Parkinson (1980) estimator for each day
sigma2_PSEI.PS <- 0.361 * (log(high / low))^2
sigma2_PSEI.PS <- na.locf(sigma2_PSEI.PS)

plot.xts(sigma2_PSEI.PS, col="black", grid.ticks.on = F)
```

### Rolling Window

```r
n <- nrow(var_data)
window <- 200
tci_values <- rep(NA, n)

for (i in window:n) {
  window_data <- var_data[(i - window + 1):i, ]

  var_window <- try(VAR(window_data, p = AIC, type = "const"), silent = TRUE)

  if (!inherits(var_window, "try-error")) {
    v_decomp <- fevd(var_window, n.ahead = 10)

    fevd_mat <- t(sapply(v_decomp, function(x) x[10, ]))


    total_spill <- (sum(fevd_mat) - sum(diag(fevd_mat))) / sum(fevd_mat) * 100
    tci_values[i] <- total_spill
  }
}

tci_series_rolling <- xts(tci_values, order.by = index(var_data))

plot.xts(tci_series_rolling,
         main = "Total volatility spillover (Rolling 200-Day Window)",
         col = "black",
         lwd = 2,
         ylab = "Spillover (in percentages)",
         grid.ticks.on=F)
```

### Expanding Window

```r

n <- nrow(var_data)
start_window <- 200
tci_values <- rep(NA, n)

for (i in start_window:n) {

  window_data <- var_data[1:i, ]

  var_window <- try(VAR(window_data, p = AIC, type = "const"), silent = TRUE)

  if (!inherits(var_window, "try-error")) {

    v_decomp <- fevd(var_window, n.ahead = 10)

    fevd_mat <- t(sapply(v_decomp, function(x) x[10, ]))

    total_spill <- (sum(fevd_mat) - sum(diag(fevd_mat))) / sum(fevd_mat) * 100

    tci_values[i] <- total_spill
  }
}

tci_series_exp <- xts(tci_values, order.by = index(var_data))

plot.xts(tci_series_exp,
         main = "Total volatility spillover (Expanding Window)",
         col = "black",
         lwd = 2,
         ylab = "Spillover (in percentages)",
         grid.ticks.on =F)
```

### Directional spillovers (rolling window)

```r
# ROLLING-WINDOW VOLATILITY SPILLOVER

n <- nrow(var_data)
window <- 200

k <- ncol(var_data)

spill_to <- matrix(NA, n, k)
spill_from <- matrix(NA, n, k)

colnames(spill_to) <- colnames(var_data)
colnames(spill_from) <- colnames(var_data)

for (i in window:n) {

  window_data <- var_data[(i-window+1):i,]

  var_model <- try(VAR(window_data, p = AIC, type = "const"), silent = TRUE)

  if (!inherits(var_model, "try-error")) {

    v_decomp <- fevd(var_model, n.ahead = 10)

    fevd_mat <- t(sapply(v_decomp, function(x) x[10,]))

    row_sums <- rowSums(fevd_mat)

    fevd_norm <- fevd_mat / row_sums

    # FROM others
    spill_from[i,] <- rowSums(fevd_norm) - diag(fevd_norm)

    # TO others
    spill_to[i,] <- colSums(fevd_norm) - diag(fevd_norm)
  }
}

spill_from_xts <- xts(spill_from * 100, order.by = index(var_data))
spill_to_xts <- xts(spill_to * 100, order.by = index(var_data))
```
